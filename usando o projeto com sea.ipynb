## Usando a blibioteca seaborn no projeto

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df_base = pd.read_csv(r'/home/gustavo82/Área de Trabalho/Projeto dados/projeto_dados1/indicadoressegurancapublicauf.csv')
```

```python
df_base.head()
```

```python
df_base = df_base[df_base['UF']=='Pará']
```

```python
df_base.head(10)
```

```python
df_base_2015 = df_base[df_base['Ano']==2015]
df_base_2016 = df_base[df_base['Ano']==2016]
df_base_2017 = df_base[df_base['Ano']==2017]
df_base_2018 = df_base[df_base['Ano']==2018]
df_base_2019 = df_base[df_base['Ano']==2019]
df_base_2020 = df_base[df_base['Ano']==2020]
df_base_2021 = df_base[df_base['Ano']==2021]
```

```python
sum_2015 = np.sum(df_base_2015['Ocorrências'])
sum_2016 = np.sum(df_base_2016['Ocorrências'])
sum_2017 = np.sum(df_base_2017['Ocorrências'])
sum_2018 = np.sum(df_base_2018['Ocorrências'])
sum_2019 = np.sum(df_base_2019['Ocorrências'])
sum_2020 = np.sum(df_base_2020['Ocorrências'])
sum_2021 = np.sum(df_base_2021['Ocorrências'])
```

```python
Ano = [2015,2016,2017,2018,2019,2020,2021]
```

```python
somas_crimes = [sum_2015,sum_2016,sum_2017,sum_2018,sum_2019,sum_2020,sum_2021]
```

```python
sns.set_style('darkgrid')
plt.subplots(figsize=(15, 8))
plt.title('Total de ocorrências no Pará de 2015 a 2021 (os dados de 2021 foram computados até junho)')
sns.barplot(x=Ano,y=somas_crimes,data = df_base)
plt.xlabel('Ano')
plt.ylabel('Total de Ocoorrências')
sns.set_style('darkgrid')
```
